# 2 Facial Landmark Detection with locations

```python
import cv2
import mediapipe as mp
from IPython.display import Image, display

def detect_landmarks(image_path):
    mp_face_mesh = mp.solutions.face_mesh
    face_mesh = mp_face_mesh.FaceMesh()

    # Function to perform landmark detection on the provided image path

    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    results = face_mesh.process(image_rgb)

    pointer_locations = []
    if results.multi_face_landmarks:
        for face_landmarks in results.multi_face_landmarks:
            for landmark in face_landmarks.landmark:
                x = int(landmark.x * image.shape[1])
                y = int(landmark.y * image.shape[0])
                pointer_locations.append((x, y))
                cv2.circle(image, (x, y), radius=2, color=(0, 255, 0), thickness=-1)

    # Display the image using IPython.display
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    display(Image(data=cv2.imencode('.jpg', image_rgb)[1].tobytes()))

    return pointer_locations
```

```python

# Specify the image path here
image_path = 'C:/Users/ASUS/OneDrive/Pictures/Saved Pictures/3.jpg'

# Perform landmark detection on the specified image path
landmark_locations = detect_landmarks(image_path)

# Print the pointer locations
for i, (x, y) in enumerate(landmark_locations):
    print(f"Pointer {i+1}: X={x}, Y={y}")
```

# 3 Facial Landmark Detection with axes and locations

```python
import cv2
import mediapipe as mp
from IPython.display import Image, display

def detect_landmarks(image_path):
    mp_face_mesh = mp.solutions.face_mesh
    face_mesh = mp_face_mesh.FaceMesh()

    # Function to perform landmark detection on the provided image path
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    results = face_mesh.process(image_rgb)

    pointer_locations = []
    if results.multi_face_landmarks:
        for face_landmarks in results.multi_face_landmarks:
            for i, landmark in enumerate(face_landmarks.landmark):
                x = int(landmark.x * image.shape[1])
                y = int(landmark.y * image.shape[0])
                pointer_locations.append((x, y))
                cv2.circle(image, (x, y), radius=2, color=(0, 0, 0), thickness=-1)
                cv2.putText(image, f"({x}, {y})",  (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.15, (0, 0, 0), 1, cv2.LINE_AA)

    # Display the image using IPython.display
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    display(Image(data=cv2.imencode('.jpg', image_rgb)[1].tobytes()))

    return pointer_locations
```

```python
# Specify the image path here
image_path = 'C:/Users/ASUS/OneDrive/Pictures/Saved Pictures/2.jpg'

# Perform landmark detection on the specified image path
landmark_locations = detect_landmarks(image_path)

# Print the pointer locations
for i, (x, y) in enumerate(landmark_locations):
    print(f"Pointer {i+1}: X={x}, Y={y}")
```
